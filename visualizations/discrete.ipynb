# Visualizations - discrete benchmarks

```python
%load_ext autoreload
%autoreload 2

import sys; sys.path.append('../')

import seaborn as sns

from evobench.discrete import Trap, StepTrap, Bimodal, StepBimodal
```

## Trap

```python
benchmark = Trap(blocks=[4])

population = benchmark.initialize_population(population_size=1e3)
sums = [solution.genome.sum() for solution in population.solutions]
benchmark.evaluate_population(population)

sns.scatterplot(x=sums, y=population.fitness)
```

## StepTrap

```python
benchmark = StepTrap(blocks=[6], step_size=2)

population = benchmark.initialize_population(population_size=1e3)
sums = [solution.genome.sum() for solution in population.solutions]
benchmark.evaluate_population(population)

sns.scatterplot(x=sums, y=population.fitness)
```
